```python
# ============================================
# Exercice 2: Base de Données Relationnelle
# ============================================
import sqlite3
import pandas as pd
from sqlalchemy import create_engine

print("=" * 60)
print("EXERCICE 2: MySQL/SQLite Database Connection")
print("=" * 60)

# ============================================
# Partie 1: Création de la base de données
# ============================================
print("\n📥 Étape 1: Chargement du fichier CSV...")

# قراءة ملف student-mat.csv
df_student = pd.read_csv('../data/student-mat.csv', sep=';')

print(f"✅ Nombre d'étudiants chargés: {len(df_student)}")
print(f"📊 Colonnes disponibles: {df_student.columns.tolist()}")

# ============================================
# Partie 2: Création de la base SQLite
# ============================================
print("\n🗄️  Étape 2: Création de la base de données SQLite...")

db_path = '../data/school_db.sqlite'
conn = sqlite3.connect(db_path)

# حفظ البيانات في جدول 'etudiants'
df_student.to_sql('etudiants', conn, if_exists='replace', index=False)

print("✅ Table 'etudiants' créée dans SQLite!")

# ============================================
# Partie 3: Requête SQL - Extraction
# ============================================
print("\n🔍 Étape 3: Exécution de la requête SQL...")
print("Requête: SELECT * FROM etudiants WHERE G3 > 15")

# الاستعلام: الطلاب اللي عندهم G3 > 15
query = "SELECT * FROM etudiants WHERE G3 > 15"
df_mysql = pd.read_sql_query(query, conn)

print(f"\n✅ Succès!")
print(f"📊 Nombre d'étudiants avec G3 > 15: {len(df_mysql)}")

print("\n📄 Aperçu des données (G1, G2, G3):")
display(df_mysql[['G1', 'G2', 'G3']].head(10))

# إضافة عمود source
df_mysql['source'] = 'MySQL'

# إغلاق الاتصال
conn.close()

print("\n✅ DataFrame MySQL créé avec succès!")
print("=" * 60)
```

```python
# ============================================
# Exercice 3: API REST
# ============================================
import requests

print("=" * 50)
print("Collecte de données via API REST")
print("=" * 50)

# الاتصال بـ API عامة (JSONPlaceholder)
api_url = "https://jsonplaceholder.typicode.com/posts"

# جلب البيانات (5 منشورات فقط)
response = requests.get(api_url, params={'_limit': 5})

if response.status_code == 200:
    # تحويل JSON إلى DataFrame
    data_api = response.json()
    df_api = pd.DataFrame(data_api)
    
    print("✅ Données récupérées avec succès!")
    print(f"\n📡 Nombre de posts: {len(df_api)}")
    print("\n📄 Aperçu des données:")
    display(df_api[['userId', 'title']])
else:
    print(f"❌ Erreur: {response.status_code}") 
```

```python
# ============================================
# Exercice 3: Web Scraping
# ============================================
from bs4 import BeautifulSoup

print("=" * 50)
print("Web Scraping - Extraction de données")
print("=" * 50)

try:
    # التحقق من robots.txt أولاً
    robots_url = "http://books.toscrape.com/robots.txt"
    robots_response = requests.get(robots_url)
    print(f"🤖 robots.txt:\n{robots_response.text[:150]}...\n")
    
    # الاتصال بالموقع
    url = "http://books.toscrape.com/"
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    
    # استخراج عناوين الكتب
    titles = [h3.a['title'] for h3 in soup.find_all('h3', class_='product_title')[:5]]
    
    print("✅ Scraping réussi!")
    print(f"\n📚 Nombre de livres extraits: {len(titles)}")
    print("\n📖 Titres des livres:")
    for i, title in enumerate(titles, 1):
        print(f"  {i}. {title}")
    
    # إنشاء DataFrame
    df_scrape = pd.DataFrame(titles, columns=['Titre_Livre'])
    print("\n📊 DataFrame créé:")
    display(df_scrape)
    
except Exception as e:
    print(f"⚠️ Erreur: {e}")
```

```python
 # ============================================
# Exercice 3: Web Scraping (Version alternative)
# ============================================
from bs4 import BeautifulSoup
import requests

print("=" * 50)
print("Web Scraping - Example")
print("=" * 50)

try:
    # استخدام موقع Wikipedia كمثال
    url = "https://en.wikipedia.org/wiki/Main_Page"
    headers = {'User-Agent': 'Mozilla/5.0'}
    page = requests.get(url, headers=headers)
    
    if page.status_code == 200:
        soup = BeautifulSoup(page.content, 'html.parser')
        
        # استخراج العناوين من الصفحة الرئيسية لـ Wikipedia
        titles = []
        for heading in soup.find_all(['h2', 'h3'])[:5]:
            text = heading.get_text(strip=True)
            if text and text != 'edit':
                titles.append(text)
        
        print("✅ Scraping réussi!")
        print(f"\n📰 Nombre d'éléments extraits: {len(titles)}")
        print("\n📋 Éléments trouvés:")
        for i, title in enumerate(titles, 1):
            print(f"  {i}. {title}")
        
        # إنشاء DataFrame
        df_scrape = pd.DataFrame(titles, columns=['Titre_Wikipedia'])
        print("\n📊 DataFrame créé:")
        display(df_scrape)
    else:
        print(f"⚠️ Erreur HTTP: {page.status_code}")
        
except Exception as e:
    print(f"⚠️ Erreur: {e}")
    print("\n💡 Solution: Créer un DataFrame exemple")
    df_scrape = pd.DataFrame({
        'Titre_Wikipedia': ['Exemple 1', 'Exemple 2', 'Exemple 3']
    })
    display(df_scrape)
```

```python
# ============================================
# FUSION DES SOURCES (Livrable final)
# ============================================

print("=" * 50)
print("FUSION MULTI-SOURCES")
print("=" * 50)

# إضافة عمود "source" لكل DataFrame
df_json['source_type'] = 'JSON'
df_excel['source_type'] = 'Excel'
df_api['source_type'] = 'API'
df_scrape['source_type'] = 'Web Scraping'

print("\n📊 Récapitulatif des sources:")
print(f"  ✓ JSON: {len(df_json)} lignes")
print(f"  ✓ Excel: {len(df_excel)} lignes")
print(f"  ✓ API: {len(df_api)} lignes")
print(f"  ✓ Web Scraping: {len(df_scrape)} lignes")

# حفظ كل DataFrame في مجلد data/
df_json.to_csv('../data/movies_processed.csv', index=False)
df_excel.to_csv('../data/donnees_processed.csv', index=False)
df_api.to_csv('../data/api_posts.csv', index=False)
df_scrape.to_csv('../data/wikipedia_scraped.csv', index=False)

print("\n✅ Tous les fichiers ont été sauvegardés dans data/!")

# إنشاء DataFrame ملخص
summary = pd.DataFrame({
    'Source': ['JSON', 'Excel', 'API', 'Web Scraping'],
    'Nombre_Lignes': [len(df_json), len(df_excel), len(df_api), len(df_scrape)]
})

print("\n📈 Résumé du TP:")
display(summary)

print("\n🎉 TP terminé avec succès!")
print("💾 N'oubliez pas de pousser sur GitHub!")
```
