```python
import pandas as pd
import numpy as np
import requests
from sqlalchemy import create_engine
import sqlite3

print("Bibliothèques chargées !")
```

```python
# 1. المصدر الأول: CSV
df_csv = pd.read_csv('../data/student-mat.csv', sep=';')

# 2. المصدر الثاني: Excel (الذي أضفته مؤخراً)
df_excel = pd.read_excel('../data/donnees_test.xlsx')

# 3. المصدر الثالث: SQL (إصلاح مشكلة الجدول المفقود)
engine = create_engine('sqlite:///../data/tp_data_mining.db')
# نقوم برفع بيانات الـ CSV للجدول أولاً لضمان وجوده
df_csv.to_sql('etudiants', engine, if_exists='replace', index=False)
df_sql = pd.read_sql("SELECT * FROM etudiants WHERE G3 > 15", engine)

# 4. المصدر الرابع: API
response = requests.get("https://jsonplaceholder.typicode.com/users")
df_api = pd.DataFrame(response.json())

print("Toutes les sources sont chargées avec succès !")
```

```python
# دمج المصادر في جدول واحد موحد
df_final = pd.concat([df_csv, df_sql, df_excel], ignore_index=True)

print("Aperçu du Pipeline Final :")
display(df_final.head())

# الإحصائيات المطلوبة في التمرين
print("\nStatistiques finales (G3) :")
print(df_final['G3'].describe())
```
